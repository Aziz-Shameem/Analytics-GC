```python
import pandas as pd
import numpy as np
import pickle
```

```python
def sigmoid(x) :
    return 1/(1 + np.exp(-x))
```

```python
model = pickle.load(open('Final Model', 'rb'))
model
```

```python
data = pd.read_csv('Final_Test.csv')
data.head()
```

```python
len(data)
```

```python
data['Prediction'] = model.predict(data)
data
```

```python
w = model.coef_.transpose()
b = model.intercept_
result = []
confidence = []
for ind in range(len(data)) :
    row = np.dot(data.iloc[ind, :4], w)
    result.append(sigmoid(row + b))
    confidence.append(abs(sigmoid(row + b)-0.5))

data['Result'] = result
data['Confidence'] = confidence
```

```python
Final = data.sort_values(by='Confidence', ascending=False).head(100)
```

```python
Final # This is the final prediction dataset
```

```python
Final.to_csv('Final_Predictions.csv')
```
